```python
import pandas as pd

# Titanic passenger details
# read csv file through github- link
data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

#to display data
data
```

```python
#To display all information about data and column names with data-type
data.info()
```

```python
# To get all static value int data of data
data.describe()
```

```python
#To check whether a value is null or not
data.isnull()
```

```python
#To count null data of each column
data.isnull().sum()
```

```python
#To fill all null (NAN) value with 1 and inplace for storing it permenant
x = data.fillna(1,inplace=True)
x
```

```python
#After the change to check whether null value exist or not to each column
data.isnull().sum()
```

```python
#To set index as Passenger Id column
new = data.set_index('PassengerId')
new
```
